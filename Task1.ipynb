```
# Write a Python program to assign grades based on marks.
# Marks	Grade
# 90 – 100	A
# 75 – 89	B
# 50 – 74	C
# Below 50	Fail
marks = int(input("Enter marks: "))

if marks >= 90 and marks <= 100:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```
# Write a Python program to create a Simple Calculator.
# The program should:
# Accept two numbers from the user.
# Accept an operator (+, -, *, /).
# Perform the selected operation and display the result.



num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
    print("Result =", result)

elif operator == "-":
    result = num1 - num2
    print("Result =", result)

elif operator == "*":
    result = num1 * num2
    print("Result =", result)

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print("Result =", result)
    else:
        print("Error: Division by zero is not allowed.")

else:
    print("Invalid Operator!")
```
